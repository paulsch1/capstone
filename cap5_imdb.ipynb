```python
import numpy as np
import pandas as pd
import glob
```

```python
name_basics = pd.read_csv('imdb/name.basics.tsv', sep='\t', low_memory=False)
title_basics = pd.read_csv('imdb/title.basics.tsv', sep='\t', low_memory=False)
title_principals = pd.read_csv('imdb/title.principals.tsv', sep='\t', low_memory=False)
title_ratings = pd.read_csv('imdb/title.ratings.tsv', sep='\t', low_memory=False)
```

```python
display(name_basics.head())
display(title_basics.head())
display(title_principals.head())
display(title_ratings.head())
```

```python
titles = pd.merge(title_basics, title_ratings, on='tconst', how='right')
```

```python
top_titles = titles.groupby('startYear').apply(lambda x: x.nlargest(10, 'numVotes')).reset_index(drop=True)
top_titles[top_titles['startYear'] == '1998']
```
