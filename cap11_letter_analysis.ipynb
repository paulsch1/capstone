# Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import string 
from nltk import FreqDist, ngrams


import warnings
warnings.filterwarnings('ignore')
```

# Load Data

```python
names_df = pd.read_csv("Data/national_data.csv")

states_df = pd.read_csv("Data/states_data.csv")
```

# Display Sample Data

```python
names_df.head(5)
```

```python
# Filter the DataFrame for male names
male_names_df = names_df[names_df['sex'] == 'M']

# Filter the DataFrame for male names
female_names_df = names_df[names_df['sex'] == 'F']
```

# N-grams Analysis:

Bigrams Analysis

```python
# Tokenize names into characters
names_df['tokens'] = names_df['name'].apply(list)

# Generate bi-grams
names_df['bigrams'] = names_df['tokens'].apply(lambda x: list(ngrams(x, 2)))

# Flatten the list of bi-grams
all_bigrams = [bg for sublist in names_df['bigrams'] for bg in sublist]

# Calculate frequency of bi-grams
bigram_freq = FreqDist(all_bigrams)

# Plot the most common bi-grams
plt.figure(figsize=(10, 6))
bigram_freq.plot(30, title='Most Common Bi-grams in Baby Names')
plt.show()
```

Get list of common bigrams

```python
# Sort the bigram_freq dictionary by frequency
sorted_bigrams = sorted(bigram_freq.items(), key=lambda x: x[1], reverse=True)

# Extract the top N most common bigrams
top_n = 10  # Change this value to the number of top bigrams you want
common_bigrams = [bigram[0] for bigram in sorted_bigrams[:top_n]]

common_bigrams
```

```python
def plot_common_bigrams(df, gender):
    # Tokenize names into characters
    df['tokens'] = df['name'].apply(list)

    # Generate bi-grams
    df['bigrams'] = df['tokens'].apply(lambda x: list(ngrams(x, 2)))

    # Flatten the list of bi-grams
    all_bigrams = [bg for sublist in df['bigrams'] for bg in sublist]

    # Calculate frequency of bi-grams
    bigram_freq = FreqDist(all_bigrams)

    # Sort the bigram_freq dictionary by frequency
    sorted_bigrams = sorted(bigram_freq.items(), key=lambda x: x[1], reverse=True)

    # Extract the top 10 most common bigrams
    top_n = 10
    common_bigrams = [bigram[0] for bigram in sorted_bigrams[:top_n]]

    # Print the list of common bigrams
    print("Top", top_n, "Most Common Bigrams for", gender, "names:")
    print(common_bigrams)

    # Plot the most common bi-grams
    plt.figure(figsize=(10, 6))
    bigram_freq.plot(top_n, title='Most Common Bi-grams in ' + gender + ' Names')
    plt.show()
```

```python
plot_common_bigrams(male_names_df, "Male")
```

```python
plot_common_bigrams(female_names_df, "Female")
```

Trigrams Analysis

```python
# Tokenize names into characters
names_df['tokens'] = names_df['name'].apply(list)

# Generate bi-grams
names_df['trigrams'] = names_df['tokens'].apply(lambda x: list(ngrams(x, 3)))

# Flatten the list of bi-grams
all_trigrams = [bg for sublist in names_df['trigrams'] for bg in sublist]

# Calculate frequency of bi-grams
trigram_freq = FreqDist(all_trigrams)

# Plot the most common bi-grams
plt.figure(figsize=(10, 6))
trigram_freq.plot(30, title='Most Common Tri-grams in Baby Names')
plt.show()
```

```python
# Sort the bigram_freq dictionary by frequency
sorted_trigrams = sorted(trigram_freq.items(), key=lambda x: x[1], reverse=True)

# Extract the top N most common bigrams
top_n = 10  # Change this value to the number of top bigrams you want
common_trigrams = [trigram[0] for trigram in sorted_trigrams[:top_n]]

common_trigrams
```

```python
def plot_common_trigrams(df, gender):
    # Tokenize names into characters
    df['tokens'] = df['name'].apply(list)

    # Generate tri-grams
    df['trigrams'] = df['tokens'].apply(lambda x: list(ngrams(x, 3)))

    # Flatten the list of tri-grams
    all_trigrams = [bg for sublist in df['trigrams'] for bg in sublist]

    # Calculate frequency of tri-grams
    trigram_freq = FreqDist(all_trigrams)

    # Sort the trigram_freq dictionary by frequency
    sorted_trigrams = sorted(trigram_freq.items(), key=lambda x: x[1], reverse=True)

    # Extract the top 10 most common trigrams
    top_n = 10
    common_trigrams = [trigram[0] for trigram in sorted_trigrams[:top_n]]

    # Print the list of common trigrams
    print("Top", top_n, "Most Common Trigrams for", gender, "names:")
    print(common_trigrams)

    # Plot the most common tri-grams
    plt.figure(figsize=(10, 6))
    trigram_freq.plot(top_n, title='Most Common Tri-grams in ' + gender + ' Names')
    plt.show()
```

```python
plot_common_trigrams(male_names_df, "Male")
```

```python
plot_common_trigrams(female_names_df, "Female")
```

# First Letter Analysis

```python
names_df['first_letter'] = names_df['name'].str[0]
letter_freq = FreqDist(names_df['first_letter'])
plt.figure(figsize=(10, 6))
letter_freq.plot(30, title='Letter Frequencies in Baby Names')
plt.show()
```

# Trends Analysis
Trends in First Letters Over Time

```python
# Create a new column for the first letter of each name
names_df['first_letter'] = names_df['name'].str[0]

# Generate the alphabet letters
alphabet_letters = list(string.ascii_uppercase)  # ['A', 'B', 'C', ..., 'Z']

# Group by year and first_letter, summing the counts
letter_trends = names_df.groupby(['year', 'first_letter'])['count'].sum().reset_index()

# Plotting trends for all letters of the alphabet
plt.figure(figsize=(12, 8))
for letter in alphabet_letters:
    if letter in letter_trends['first_letter'].unique():
        filtered_trends = letter_trends[letter_trends['first_letter'] == letter]
        plt.plot(filtered_trends['year'], filtered_trends['count'], label=letter)

plt.title('Trends in Letters in Baby Names')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='First Letter', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Trends in First Letters of Male Names

```python
# Create a new column for the first letter of each name (converted to lowercase)
male_names_df['first_letter'] = male_names_df['name'].str[0].str.lower()

# Generate the lowercase alphabet letters
alphabet_letters = list(string.ascii_lowercase)  # ['a', 'b', 'c', ..., 'z']

# Group by year and first_letter, summing the counts
letter_trends_male = male_names_df.groupby(['year', 'first_letter'])['count'].sum().reset_index()

# Plotting trends for all letters of the alphabet for male names
plt.figure(figsize=(12, 8))
for letter in alphabet_letters:
    if letter in letter_trends_male['first_letter'].unique():
        filtered_trends = letter_trends_male[letter_trends_male['first_letter'] == letter]
        plt.plot(filtered_trends['year'], filtered_trends['count'], label=letter.upper())

plt.title('Trends in First Letters of Male Names')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='First Letter', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
