```python
import numpy as np
import pandas as pd

from data_loader import load_names_from_web, holdout_split
```

```python
dfraw = load_names_from_web(category='national', hide_pre_1937=False, use_existing_files=True)   # default category="both", hide_pre_1937=True
dfraw.sample(20)
```

```python
traintestval, holdout = holdout_split(dfraw)
display(traintestval.sort_values(by='count', ascending=False).head(20))
display(holdout.sort_values(by='count', ascending=False).head(20))
display(len(traintestval))
display(len(holdout))
```

```python
trainval, test = holdout_split(traintestval)
display(trainval.sort_values(by='count', ascending=False).head(20))
display(test.sort_values(by='count', ascending=False).head(20))
display(len(trainval))
display(len(test))
```
